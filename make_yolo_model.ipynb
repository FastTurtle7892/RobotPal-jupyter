```python
from ultralytics import YOLO
import os

# ==========================================
# 1. 설정
# ==========================================
# 데이터셋 폴더 이름 (본인이 압축 푼 폴더 이름으로 바꾸세요!)
dataset_folder_name = "license_plate_dataset" 

# data.yaml 파일의 절대 경로를 자동으로 찾습니다
current_dir = os.getcwd()
data_yaml_path = os.path.join(current_dir, dataset_folder_name, "data.yaml")

print(f"데이터셋 경로: {data_yaml_path}")

# ==========================================
# 2. 모델 생성 및 학습
# ==========================================
# 가장 가벼운 모델(yolov8n.pt)을 베이스로 시작
model = YOLO("yolov8n.pt") 

print(">>> 학습을 시작합니다... (시간이 좀 걸립니다)")

# 학습 시작!
# imgsz=416: Jetson Nano처럼 성능이 낮은 기기에서는 416이나 320 추천
# epochs=50: 50번 반복 학습
model.train(data=data_yaml_path, epochs=100, imgsz=416)

print(">>> 학습 완료!")
```
