```python
import ipywidgets
from IPython.display import display
import robotpal.robot_config as cfg
import robotpal.robot_vision as Rvision
import robotpal.robot_logic as Rlogic
import time

# 1. 하드웨어 초기화
cfg.init_servos()
```

```python
# 2. 비전 시스템 및 컨트롤러 생성
vision_sys = Rvision.VisionSystem()
controller = Rlogic.DrivingController(vision_sys)

# 3. 감지 스레드 시작
vision_sys.start_detection_thread()

# UI 위젯 배치
ui_layout = ipywidgets.HBox([
    cfg.speed_slider, 
    cfg.steering_slider
])

display(cfg.image_widget)
display(ui_layout)
display(cfg.log_widget)

# 카메라 연결 및 실행 시작
try: 
    cfg.camera.unobserve_all()
except: 
    pass

# 초기 실행 및 이벤트 연결
controller.execute({'new': cfg.camera.value})
cfg.camera.observe(controller.execute, names='value')

cfg.log_print(">>> 시스템 가동 시작!")
```

```python

cfg.camera.unobserve(controller.execute, names='value')
time.sleep(0.1)

# 스레드 종료
vision_sys.stop_detection_thread()

# 로봇 정지
cfg.robot.stop()
cfg.camera.stop()

cfg.log_print(">>> 시스템 종료됨")
```
