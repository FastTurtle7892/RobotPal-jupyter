### Resolved Motion Control

로봇 팔은 두 개의 서보모터를 이용해서 위치 제어를 하므로, 두 서보모터를 각각 제어하면, 위치 제어가 어렵습니다.  
그래서 다음과 같은 API 를 사용합니다.  
xyInput(x,y) : 단위는 mm 이며, X값이 클수록, 팔의 위치가 앞쪽으로 이동합니다. Y값이 클수록 팔의 위치가 높아집니다.

```python
TTLServo.xyInput(100, 0)
```

X와 Y 값의 관계를 더 직관적으로 보여주기 위해, 로봇 팔을 (100, 0)에서 (200, 0)으로 단계별로 이동한 다음 다시 위치 (150, 0)으로 이동시킵니다.  
여기서의 이동은 위치의 X 매개변수를 변경합니다.  
로봇 팔의 움직임을 관찰하면서 다음 코드를 실행하세요.

```python
import time

# （100，0）이동
TTLServo.xyInput(100, 0)
time.sleep(1)

# （100，0）에서（200，0）까지 이동
for i in range(100, 201):
    TTLServo.xyInput(i, 0)
    time.sleep(0.01)
time.sleep(1)

# （100，0）에서（100，100）까지 왕복하는 부드러운 이동이며, 세 번째 매개변수 값은 해당 위치로 이동하는 시간이며 단위는 초 입니다.

for i in range(0,4):
    TTLServo.xyInputSmooth(100, 0, 1)
    time.sleep(2)
    TTLServo.xyInputSmooth(100, 100, 1)
    time.sleep(2)

# （150，0）이동
TTLServo.xyInput(150, 0)
time.sleep(1)
```

다음 코드는 Y 값만 변경합니다.  
로봇 팔의 움직임을 관찰하면서 다음 코드를 실행하세요.

```python
# （150，-50）이동
TTLServo.xyInput(150, -50)
time.sleep(1)

# ( 150，-50）에서（150，50）까지 이동
 for i in range(-50, 51):
    TTLServo.xyInput(150, i)
    time.sleep(0.01)
time.sleep(1)

# （150， 0）이동
TTLServo.xyInput(150, 0)
time.sleep(1)
```

서보모터를 사용할 때, 때때로 회전을 시작하고 회전 각도가 결정되지 않는 경우가 있습니다.  
예를 들어, servoAngleCtrl(1, 80, 1, 150)을 사용하여 카메라를 오른쪽으로 회전시키고, 그 후 servoStop(1)를 호출하여 1번 서보를 정지할 수 있습니다.  
여기서 80은 제한 각도를 설정하는 데에도 사용할 수 있습니다.

```python
TTLServo.servoAngleCtrl(1, 0, 1, 150)
time.sleep(1)
TTLServo.servoStop(1)
```
